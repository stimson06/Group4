Sort all images into Unique sub-folders. Make life easier and looks cleaner and organised

```python
import os
import shutil

folder_path = r"D:\Projects\Group4\handset_dataset" # Add whatever path you are storing the dataset at 

classes = ["backward", "backwards", "forward", "left", "right", "unknown"] # Hope everyone added unknown classes

for c in classes:
    os.makedirs(os.path.join(folder_path, c), exist_ok=True)

moved = 0
skipped = 0

for filename in os.listdir(folder_path):
    filepath = os.path.join(folder_path, filename)
    if not os.path.isfile(filepath):
        continue

    prefix = filename.split("_")[0]
    if prefix in classes:
        dest = os.path.join(folder_path, prefix, filename)
        shutil.move(filepath, dest)
        moved += 1
    else:
        skipped += 1

print(f"Done. Moved: {moved} | Skipped: {skipped}")
```
